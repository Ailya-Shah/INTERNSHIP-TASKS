## 2. The question: decision, action, cost of a wrong call

**Decision:** which of thousands of content pages an editor with limited hours should review first for page-one potential.

**Action:** an editor reviews/edits the highest-ranked pages first — expanding thin content, refreshing stale pages, tightening query focus — based on the signals my analysis flags.

**Cost of a wrong call:** if the model flags a page as "worth reviewing" and it isn't, that's wasted editor hours — annoying but recoverable. If it misses a page that actually would move to page one with a small edit, that's a missed opportunity, silent and harder to notice. Because both costs are moderate (not safety-critical), I don't need a perfect model — a *ranking* that's meaningfully better than guessing already helps.

**Why data/ML helps at all:** with one signal (say, word count alone) a human could write a simple rule. But page-one status plausibly depends on several tangled signals at once (length, freshness, demand, query concentration) — that's exactly the situation where a simple learned model earns its keep over a hand-written if-statement.


## 3. Quick look at the data (2-3 real numbers)

Loading the starter CSV to show why this lane looks worth the next 7 weeks.


```python
import pandas as pd

df = pd.read_csv("../../data/raw/content_refresh_anonymized.csv")
print(f"rows: {len(df):,} | clients: {df['client_id'].nunique()}")

# avg_position == 0 means "no data" in this dataset, not literal position zero -- exclude it
valid = df[df["avg_position"] > 0].copy()
valid["is_page_one"] = valid["avg_position"].between(1, 10)

base_rate = valid["is_page_one"].mean()
print(f"page-one base rate: {base_rate:.1%}  (n={len(valid):,} pages with real position data)")

med_p1 = valid.loc[valid.is_page_one, "word_count"].median()
med_other = valid.loc[~valid.is_page_one, "word_count"].median()
print(f"median word_count -- page-one: {med_p1:.0f}  vs  not page-one: {med_other:.0f}")
```
